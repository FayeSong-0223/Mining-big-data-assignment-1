### 1. Read the dataset

```python
import pandas as pd

# Read the CSV data
df = pd.read_csv('ObesityDataSet.csv')

# Display the first 5 rows
print("Dataset Preview:")
print(df.head())

# Display basic dataset information
print("\nDataset Information:")
print(df.info())

# Display descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())
```

### 2. Analyse and visualise the data

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Plot the distribution of the target variable NObeyesdad
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='NObeyesdad', order=df['NObeyesdad'].value_counts().index)
plt.title("Distribution of Obesity Categories")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# If there are numerical variables like Age, Height, and Weight, plot their distributions
num_features = ['Age', 'Height', 'Weight'] if set(['Age','Height','Weight']).issubset(df.columns) else df.select_dtypes(include='number').columns.tolist()
df[num_features].hist(bins=20, figsize=(12,8))
plt.suptitle("Distribution of Numerical Features")
plt.show()

# Plot a correlation heatmap for numerical variables
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### 3. Organise the data for modelling

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Create a copy of the dataset to avoid altering the original
data = df.copy()

# If Height and Weight exist, calculate BMI (BMI = Weight(kg) / (Height(m))^2)
if set(['Height','Weight']).issubset(data.columns):
    # Convert Height from cm to m
    data['Height_m'] = data['Height'] / 100.0
    data['BMI'] = data['Weight'] / (data['Height_m'] ** 2)
    print("New BMI column added:")
    print(data[['Height', 'Weight', 'BMI']].head())

# Encode the target variable NObeyesdad
le_target = LabelEncoder()
data['NObeyesdad_enc'] = le_target.fit_transform(data['NObeyesdad'])

# Identify categorical features for one-hot encoding (if they exist)
# For example: Gender, FAVC, CAEC, MTRANS
categorical_features = []
for col in ['Gender', 'FAVC', 'CAEC', 'MTRANS']:
    if col in data.columns:
        categorical_features.append(col)

# Apply one-hot encoding to these categorical variables
data_encoded = pd.get_dummies(data, columns=categorical_features, drop_first=True)

# Create the feature matrix X and target variable y (using NObeyesdad_enc as target)
X = data_encoded.drop(['NObeyesdad', 'NObeyesdad_enc'], axis=1)
y = data_encoded['NObeyesdad_enc']

# Scale numerical features using StandardScaler
scaler = StandardScaler()
# Note: We typically do not scale dummy variables, so we apply scaling only to numerical columns
num_cols = X.select_dtypes(include=np.number).columns
X[num_cols] = scaler.fit_transform(X[num_cols])

print("\nSample of preprocessed features:")
print(X.head())
```

### 4. Implement prediction models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Instantiate the models
model_lr = LogisticRegression(max_iter=1000, random_state=42)
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)

print("Two candidate models have been constructed: Logistic Regression and Random Forest.")
```

### 5. Train prediction models

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# --- Ensure that the feature matrix X contains only numeric values ---
# Print any columns with object type in X (for debugging)
print("Object-type columns in X before additional encoding:")
print(X.select_dtypes(include=['object']).columns)

# If there are any non-numeric columns, apply one-hot encoding on them
if len(X.select_dtypes(include=['object']).columns) > 0:
    X = pd.get_dummies(X, drop_first=True)
    print("After get_dummies, object-type columns in X:")
    print(X.select_dtypes(include=['object']).columns)

# --- Split the data into training and testing sets ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Confirm that X_train contains only numeric columns
print("Object-type columns in X_train after splitting:")
print(X_train.select_dtypes(include=['object']).columns)

# --- Initialize the models ---
model_lr = LogisticRegression(max_iter=1000, random_state=42)
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)

# --- Train the models ---
model_lr.fit(X_train, y_train)
model_rf.fit(X_train, y_train)

print("Both models have been successfully trained.")
```

### 6. Test prediction models and show results

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np

# Predictions and evaluation for Logistic Regression
y_pred_lr = model_lr.predict(X_test)
acc_lr = accuracy_score(y_test, y_pred_lr)
print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr))
print("Logistic Regression Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_lr))
print("Logistic Regression Accuracy:", np.round(acc_lr, 4))

# Predictions and evaluation for Random Forest
y_pred_rf = model_rf.predict(X_test)
acc_rf = accuracy_score(y_test, y_pred_rf)
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
print("Random Forest Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
print("Random Forest Accuracy:", np.round(acc_rf, 4))

```

### 7. Compare the results from all candidate models, choose the best model, justify your choice and discuss the results

```python
import pandas as pd

# Create a summary table of results
results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest'],
    'Accuracy': [acc_lr, acc_rf]
})
print("\nComparison of Model Performance:")
print(results)

# If Random Forest performs better, display its feature importances
if acc_rf >= acc_lr:
    importances = model_rf.feature_importances_
    feat_names = X.columns
    feature_importance_df = pd.DataFrame({'Feature': feat_names, 'Importance': importances})
    feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
    
    print("\nFeature Importances from Random Forest:")
    print(feature_importance_df)
    
    # Plot the top 10 most important features
    plt.figure(figsize=(10,6))
    sns.barplot(data=feature_importance_df.head(10), x='Importance', y='Feature')
    plt.title("Top 10 Important Features")
    plt.tight_layout()
    plt.show()
    
    best_model = "Random Forest"
else:
    best_model = "Logistic Regression"

print("\nSelected Best Model:", best_model)

```
